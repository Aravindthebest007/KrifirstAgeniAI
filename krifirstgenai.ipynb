```python
import re

def parse_rubric_ratings(response_text):
    """
    Parses the text response from the generative model (markdown table format)
    to extract rubric ratings.

    Args:
        response_text: The text response from the generative model in markdown table format.

    Returns:
        A dictionary where keys are Product IDs and values are dictionaries
        of rubric ratings.
    """
    ratings = {}
    lines = response_text.strip().split('\n')

    # Find the start of the ratings data (skip header and separator lines)
    start_index = -1
    for i, line in enumerate(lines):
        if line.startswith("| Product ID"):
            start_index = i + 2 # Skip header and separator line
            break

    if start_index == -1:
        print("Could not find the start of the ratings data in the response.")
        return ratings

    # Process lines containing product ratings
    for line in lines[start_index:]:
        # Split the line by the '|' delimiter and remove leading/trailing whitespace
        parts = [part.strip() for part in line.split('|')]
        # Ensure there are enough parts and the first part is empty (due to leading '|')
        if len(parts) >= 5 and parts[0] == '':
            product_id = parts[1]
            try:
                style_rating = int(parts[2])
                suitability_rating = int(parts[3])
                customer_reviews_rating = int(parts[4])

                ratings[product_id] = {
                    'Style/Trendiness': style_rating,
                    'Suitability for London Weather': suitability_rating,
                    'Customer Reviews': customer_reviews_rating
                }
            except ValueError:
                # Handle cases where ratings are not valid integers
                print(f"Skipping line due to invalid rating: {line}")
                continue

    return ratings

# Call the function with the response and store the parsed ratings
parsed_ratings = parse_rubric_ratings(response)

print("Parsed Rubric Ratings:")
print(parsed_ratings)
```
